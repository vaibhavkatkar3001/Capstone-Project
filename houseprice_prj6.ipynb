# KNN - Predict the Price of Bangalore House



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
```

```python
data =pd.read_csv("C:/Users/soura/OneDrive/Desktop/Vaibhav DS/vibs project/Capstone project 1/Capstone Project  - 1 KNN Prj_5&Prj_6/KNN Assignment/bangalore house price prediction OHE-data.csv")
```

```python
print(data.head(2))
```

```python
data = pd.DataFrame(data)
data
```

```python
y = data['price']
X = data.drop('price',axis = 1)
```

```python
print(y.shape)
```

```python
print(X.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 0 )
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
math.sqrt(len(y_test))
```

```python
model = KNeighborsRegressor(n_neighbors=5)
model.fit(X_train,y_train)
```

```python
model_fit = model.fit(X_train,y_train)
```

```python
round(model_fit.score(X_train,y_train),2)
```

```python
y_pred = model_fit.predict(X_test)
results = r2_score(y_test,y_pred)
print(results)
```

```python
error = []
k = []
for i in range(1,21):
    model = KNeighborsRegressor(n_neighbors=i)
    model_fit = model.fit(X_train,y_train)
    err = 1 - round(model_fit.score(X_train, y_train),2)
    error.append(err)
    k.append(i)
```

```python
pd.DataFrame({'k':k, 'error':error})
```
